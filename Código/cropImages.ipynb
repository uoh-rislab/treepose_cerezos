Función para recortar imágenes

```python
#Se importan librerias
import os
import cv2
import numpy as np

#Funcion Pricipal
def main():
    #Directorios
    path_folder = '/content/drive/MyDrive/Universidad/Tesis/Datasets/raw'
    path_saving = '/content/drive/MyDrive/Universidad/Tesis/Datasets/cropped'
    list_folders = [file for file in sorted(os.listdir(path_folder))]

    #Crea directorio de guardado si no existe
    if not os.path.isdir(path_saving):
        os.mkdir(path_saving)

    #Se define particion
    partition = [1,3]
    assert (len(partition)==2),'ERROR INVALID PARTITION'
    img1 = cv2.imread(os.path.join(path_folder, list_folders[0]))
    shape = np.array(img1.shape[:2],dtype=np.int32)
    delta_y = shape[0] // partition[0]
    delta_x = shape[1] // partition[1]
    ext = '.JPG'

    #Se recorta y guarda
    for img_name in list_folders:
        path_img = os.path.join(path_folder, img_name)
        if not (ext in img_name):
            print(img_name)
            continue
        img = cv2.imread(path_img)

        for i in range(partition[0]):
            ylim = [delta_y * i, delta_y * (i + 1)]
            for j in range(partition[1]):
                xlim = [delta_x * j, delta_x * (j + 1)]
                cut_img = img[ylim[0]:ylim[1], xlim[0]:xlim[1], :]
                sufix = '_{:d}{:d}{:s}'.format(i, j, ext)
                new_name = img_name.replace(ext, sufix)
                new_path = os.path.join(path_saving, new_name)
                cv2.imwrite(new_path, cut_img)

main()
```
